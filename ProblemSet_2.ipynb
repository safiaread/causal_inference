```python
import pandas as pd
import numpy as np
nsw= pd.read_csv("Downloads/nsw_dw-1.csv")
cps = pd.read_csv("Downloads/cps_dw-1.csv")
```

```python
#Keely and I collaborated.

# 1.
#The difference in means is an unbiased estimator for the ATE when no hidden treatment, unconditional unconfoundedness, 
#and no interference are true. No interence means the assignemnt of job training or not does not affect income in 
#1973. No hidden treatment means the people who are assigned to job training go to job training and the ones who 
#are not, do not. Unconditional unconfoundedness means there are no covariates that affect both the assignment of 
#job training and 1973 income. 

nsw[nsw.treat == 1].re78.mean()-nsw[nsw.treat == 0].re78.mean()
```

```python
#2. 
#Most of the covariates are balanced, but no degree and hispanic have 
#larger differences than the others. 
#This would cause bias in the estimate and a subclassification estimator 
#may be more effective (DiCM).
nsw_mean = nsw[["treat","age", "educ", "black", "hisp", "marr", "nodegree", "re74", "re75", "unemp74", "unemp75"]].groupby("treat").mean()
nsw_sd = nsw[["treat","age", "educ", "black", "hisp", "marr", "nodegree", "re74", "re75", "unemp74", "unemp75"]].groupby("treat").std()
nsw_mean.join(nsw_sd, lsuffix='_mean', rsuffix='_sd')
```

```python
#3.
cat = pd.concat([nsw,cps])
cat
```

```python
#4. 
print("The DiM is", cat[cat.treat == 1].re78.mean()-cat[cat.treat == 0].re78.mean())

cat_mean = cat[["treat","age", "educ", "black", "hisp", "marr", "nodegree", "re74", "re75", "unemp74", "unemp75"]].groupby("treat").mean()
cat_sd = cat[["treat","age", "educ", "black", "hisp", "marr", "nodegree", "re74", "re75", "unemp74", "unemp75"]].groupby("treat").std()
cat_mean.join(cat_sd, lsuffix='_mean', rsuffix='_sd')

#The covariate balance is much less balanced in the combined dataset. 
#A subclassification estimator would be more accurate.
```

```python
#5.

x1 = cat[(cat["nodegree"] ==0) & (cat["hisp"]==0) & (cat["black"]==0)]
x2 = cat[(cat["nodegree"] ==0) & (cat["hisp"]==0) & (cat["black"]==1)]
x3 = cat[(cat["nodegree"] ==0) & (cat["hisp"]==1) & (cat["black"]==1)]
x4 = cat[(cat["nodegree"] ==1) & (cat["hisp"]==1) & (cat["black"]==1)]
x5 = cat[(cat["nodegree"] ==0) & (cat["hisp"]==1) & (cat["black"]==0)]
x6 = cat[(cat["nodegree"] ==1) & (cat["hisp"]==1) & (cat["black"]==0)]
x7 = cat[(cat["nodegree"] ==1) & (cat["hisp"]==0) & (cat["black"]==1)]
x8 = cat[(cat["nodegree"] ==1) & (cat["hisp"]==0) & (cat["black"]==0)]

px1= x1.shape[0]/cat.shape[0]
px2= x2.shape[0]/cat.shape[0]
px3= x3.shape[0]/cat.shape[0]
px4= x4.shape[0]/cat.shape[0]
px5= x5.shape[0]/cat.shape[0]
px6= x6.shape[0]/cat.shape[0]
px7= x7.shape[0]/cat.shape[0]
px8= x8.shape[0]/cat.shape[0]

cate1= x1[x1["treat"]==1].re78.mean() - x1[x1["treat"]==0].re78.mean()
cate2= x2[x2["treat"]==1].re78.mean() - x2[x2["treat"]==0].re78.mean()
cate3= x3[x3["treat"]==1].re78.mean() - x3[x3["treat"]==0].re78.mean()
cate4= x4[x4["treat"]==1].re78.mean() - x4[x4["treat"]==0].re78.mean()
cate5= x5[x5["treat"]==1].re78.mean() - x5[x5["treat"]==0].re78.mean()
cate6= x6[x6["treat"]==1].re78.mean() - x6[x6["treat"]==0].re78.mean()
cate7= x7[x7["treat"]==1].re78.mean() - x7[x7["treat"]==0].re78.mean()
cate8= x8[x8["treat"]==1].re78.mean() - x8[x8["treat"]==0].re78.mean()

(px1*cate1)+(px2*cate2)+(px5*cate5)+(px6*cate6)+(px7*cate7)+(px8*cate8)
#x3 and x4 (with both hispanic and black as a part of the subset) 
#was removed due to not having any individuals in the subset.

#This value is similar to the DiM in Question 4, but the estimated 
#effect is a little weaker.
```

```python
#6.
# Preparing med datafram:
# 1
T = 'treat'
Y = 're78'

cat['agesq'] = cat['age']*cat['age']
cat['agecub'] = cat['age']*cat['age']*cat['age']
cat['educsq'] = cat['educ']*cat['educ']
cat['educxre74'] = cat['educ']*cat['re74'] 

#2
X = ['age', 'agesq', 'agecub', 'educ', 'educsq', 'marr', 'nodegree', 'black', 'hisp', 're74','educxre74','unemp74','unemp75']


from sklearn.linear_model import LogisticRegression as LR

#3
ps_model = LR().fit(cat[X], cat[T])

#4
data_ps_1 = cat.assign(e_hat=ps_model.predict_proba(cat[X])[:, 1])

print(data_ps_1)

e_hat = data_ps_1['e_hat']

#We include higher order and interaction terms because some of the 
#covariates affect each other. For instance you may be more likely to 
#have more education if you are older. 
```

```python
#7. 
from sklearn.neighbors import KNeighborsRegressor

treated = data_ps_1.query("treat==1")
untreated = data_ps_1.query("treat==0")

X = ['e_hat']
y = 're78'

mt0 = KNeighborsRegressor(n_neighbors=1).fit(untreated[X], untreated[y])
mt1 = KNeighborsRegressor(n_neighbors=1).fit(treated[X], treated[y])

predicted = pd.concat([
    # find matches for the treated looking at the untreated knn model
    treated.assign(match=mt0.predict(treated[X])),
    
    # find matches for the untreated looking at the treated knn model
    untreated.assign(match=mt1.predict(untreated[X]))
])

predicted.head()
np.mean((2*predicted["treat"] - 1)*(predicted["re78"] - predicted["match"]))

#For the Propensity Score Matching Estimator to be unbiased, we would assume that we 
#know/are able to model the P(T=1|X), X in independent from T|P(T=1|X), and the potential 
#outcomes Y(1),Y(0) are independent from T|P(T=1|X), and that there is strict overlap and 
#the number of observed treated individuals is not 0 or the entire set of observations.
#Because the study is a randomized control trial, it should satisfy these assumptions.
```
