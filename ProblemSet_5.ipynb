```python
import pandas as pd
import numpy as np
card = pd.read_csv("Downloads/CardKrueger.csv")
card.head()
```

```python
#Question 1

card['empfte'] = card['empft']+0.5*card['emppt']+card['nmgrs']
card['empfte2'] = card['empft2']+0.5*card['emppt2']+card['nmgrs2']
card
```

```python
#Question 2

grouped = card.groupby('state')

# Applying a function to each group, for example, calculating the mean
mean_employ_before = grouped['empfte'].mean()
mean_employ_before
mean_employ_after = grouped['empfte2'].mean()
mean_employ_after
merged_df = pd.merge(mean_employ_before, mean_employ_after, on='state', how='left')
merged_df
mean_employ_after-mean_employ_before

#We found employment increased by 0.588021 in New Jersey and decreased by -2.165584 in Pennsylvania. 
#These measures don't control for time variant factors such as inflation that might confound the estimate so are flawed estimates.
```

```python
#Question 3

(mean_employ_after[1]-mean_employ_before[1])-(mean_employ_after[2]-mean_employ_before[2]) 
```

```python
#Question 4
card['change'] = card['empfte2']-card['empfte']

nj = card[card['state'] == 1]
penn = card[card['state'] == 2]

from scipy import stats

# Perform independent t-test
t_statistic, p_value = stats.ttest_ind(nj['change'], penn['change'], nan_policy='omit')

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

#It is statistically significant at a threshold of p < 0.05. 

card
```

```python
#Question 6 

card['i'] = card.index
reshaped = card[['i','empfte', 'empfte2', 'state']]
reshaped.loc[:, 'nj'] = np.where(reshaped['state'] == 1, 1, 0)

reshaped = reshaped.drop('state', axis=1)

reshaped

# Pivot longer using melt()
reshaped_2 = pd.melt(reshaped, id_vars=['i','nj'], var_name='Post', value_name='Y')

reshaped_2['Post'] = np.where(reshaped_2['Post'] == "empfte", 0, 1)

print(reshaped_2)
```

```python
#Question 7

import statsmodels.api as sm

model = sm.formula.ols(formula='Y ~ nj + Post + nj:Post', data=reshaped_2).fit()

print(model.summary())

#The DD estimate is 2.7536. Yes, the regression coefficient of our interaction term is the 
#same as our result from Question 3. The standard error is 1.688 for the interaction term.
```

```python
#Question 8

#The P Value of the interaction term is 0.103 which is greater than the typical P Value of 
#0.05, so it is not significant. 
#The P Value in Question 4 is likely less accurate because it doesn't control for 
#confounders, unlike this regression which has an interaction term.
```
